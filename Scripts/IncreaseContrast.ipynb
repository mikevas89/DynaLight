```python
import cv2
import numpy as np
from PIL import Image
import scipy
import scipy.fftpack
from pylab import plot, show, title, xlabel, ylabel, subplot
from matplotlib import pyplot as plt
import pylab
from scipy import fft, arange, signal
import numpy


%matplotlib inline

file_name = "/home/mike/Documents/Experiments/distanceComparison/big_led/20cm/1500.jpg"

gray_image = cv2.imread(file_name, cv2.IMREAD_GRAYSCALE)
image = cv2.imread(file_name)
plt.imshow(gray_image, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.title('OriginalGray')
plt.show()

print "rows:", gray_image.shape[0], "columns:", gray_image.shape[1]

gray_middle_pixel_values = []
#for every row 
for i in range (0, gray_image.shape[0]):
    #get the middle column pixel value
    #print blur[i,gray_image.shape[1]/2-1]
    gray_middle_pixel_values.append(gray_image[i,gray_image.shape[1]/2-1])

print sum(gray_middle_pixel_values) / float(len(gray_middle_pixel_values))
```

```python
equ = cv2.equalizeHist(gray_image)

res = np.hstack((gray_image,equ)) #stacking images side-by-side
plt.imshow(res, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.title('Original and equalize1')
plt.show()

# create a CLAHE object (Arguments are optional).
clahe = cv2.createCLAHE(clipLimit=10.0, tileGridSize=(8,8))
cl1 = clahe.apply(gray_image)

res = np.hstack((gray_image,cl1)) #stacking images side-by-side
plt.imshow(res, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.title('Original and equalize2')
plt.show()


```

```python

blur1 = cv2.blur(equ,(50,50))
res1 = np.hstack((gray_image,equ, blur1)) #stacking images side-by-side
plt.imshow(res1, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.title('OriginalGray equ1 Blur1')
plt.show()

blur2 = cv2.GaussianBlur(equ,(11,11),0)
res2 = np.hstack((gray_image,equ, blur2)) #stacking images side-by-side
plt.imshow(res2, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.title('OriginalGray equ1 Blur2')
plt.show()

median = cv2.medianBlur(equ,5)
res3 = np.hstack((gray_image,equ, median)) #stacking images side-by-side
plt.imshow(res3, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.title('OriginalGray equ1 Blur3')
plt.show()

```

```python
middle_pixels = []

#for every row 
for i in range (0, blur2.shape[0]):
    #get the middle column pixel value
    #print blur2[i,gray_image.shape[1]/2-1]
    middle_pixels.append(blur1[i,gray_image.shape[1]/2-1])
    
#plot blur pixels
plt.plot(middle_pixels)
plt.ylabel('blur pixels')
plt.title('MiddlePixels Blur1')
plt.show()

average_pixels = []
x = 0
#for every row 
for i in range (0, blur1.shape[0]):
    for j in range (0, blur2.shape[1]):
             x = x + blur1[i,j]       
    average_pixels.append(x/blur1.shape[1])
    x = 0

#plot blur pixels
plt.plot(average_pixels)
plt.ylabel('blur pixels')
plt.title('Average Pixels Blur1')
plt.show()
```

```python

#plot blur pixels
plt.plot(average_pixels)
plt.ylabel('blur pixels')
plt.title('Average Pixels Blur1')
plt.show()


window_size = 100
window = numpy.ones(int(window_size))/float(window_size)
new =  numpy.convolve(average_pixels, window, 'same')

plt.plot(new)
plt.title('Rolling Window')
plt.ylabel('blur pixels')
plt.show()

peaks = []
falls = []
for i in range (1, new.shape[0]-1):
    if(new[i-1] < new[i] and new[i+1] < new[i]):
        #print "peak", new[i]
        peaks.append(new[i])
    if(new[i-1] > new[i] and new[i+1] > new[i]):
        #print "fall" , new[i]
        falls.append(new[i])

averagePeak = sum(peaks)/len(peaks)
averageFall = sum(falls)/len(falls)
minPeak = min(peaks)
maxFall = max(falls)

print averagePeak , averageFall
averagePeakThreshold = (averageFall + averagePeak) / 2
print "AveragePeakThresh", averagePeakThreshold

minMaxThreshold = (minPeak + maxFall) / 2
print "MinMaxThreshold: ", minMaxThreshold

averageThreshold = sum(new)/len(new)
print "AverageThreshold", averageThreshold

threshold = averageThreshold
final1 = []
for i in range (0, blur2.shape[0]):
    if(new[i]>threshold):
        final1.append(1)
    else:
        final1.append(0)

plt.plot(final1)
plt.ylabel('blur pixels')
plt.ylim([-1,2])

plt.show()
```
