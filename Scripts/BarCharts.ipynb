```python
"""
Bar chart demo with pairs of bars grouped for easy comparison.
"""
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


n_groups = 3

packet1 = (1, 94, 98)

packet2 = (1, 84, 100)

packet3 = (1, 66, 96)

fig, ax = plt.subplots()

index = np.arange(n_groups)
bar_width = 0.15

opacity = 0.9


rects1 = plt.bar(index + bar_width/2 , packet1, bar_width, alpha=opacity,
                 color='#4d9900',
                 label='Packet1')

rects2 = plt.bar(index + 3*bar_width/2, packet2, bar_width, alpha=opacity,
                 color='#538cc6',
                 label='Packet2')

rects3 = plt.bar(index + (5*bar_width/2), packet3, bar_width, alpha=opacity,
                 color='#ffdb4d',
                 label='Packet3')




plt.xlabel('Pixel Selection Method')
plt.ylabel('Accuracy (%)')
plt.xticks(index + 2*bar_width, ('MiddleColumn', 'Center Offset', 'Offset'))
plt.yticks(np.arange(0, 101, 10.0))
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),
          fancybox=True, shadow=True, ncol=5)
plt.tight_layout()
plt.show()
```
