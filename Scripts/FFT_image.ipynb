```python
import cv2
import numpy as np
from PIL import Image
import scipy
import scipy.fftpack
from pylab import plot, show, title, xlabel, ylabel, subplot
from matplotlib import pyplot as plt
import pylab
from scipy import fft, arange, signal
import numpy


%matplotlib inline

file_name = "/home/mike/Documents/sample200hz/-20.jpg"

gray_image = cv2.imread(file_name, cv2.IMREAD_GRAYSCALE)
'''plt.imshow(gray_image, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
'''
image = cv2.imread(file_name)
b,g,r = cv2.split(image)       # get b,g,r
rgb_img = cv2.merge([r,g,b])     # switch it to rgb

blur = cv2.blur(gray_image,(50,50))
'''
plt.imshow(blur, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
'''
print "rows:", gray_image.shape[0], "columns:", gray_image.shape[1]

blur_middle_pixel_values = []
gray_middle_pixel_values = []

#for every row 
for i in range (0, gray_image.shape[0]):
    #get the middle column pixel value
    #print blur[i,gray_image.shape[1]/2-1]
    blur_middle_pixel_values.append(blur[i,gray_image.shape[1]/2-1])
    gray_middle_pixel_values.append(gray_image[i,gray_image.shape[1]/2-1])

#print gray_middle_pixel_values



mean = numpy.mean(gray_middle_pixel_values)
print mean

withoutDC = gray_middle_pixel_values - mean

y = withoutDC * numpy.hamming(withoutDC.shape[0])

```

```python
# FFT calculations

camera_rolling_shutter_r = 1/32000.0

y = withoutDC
L = len(y)

Fs = 1/camera_rolling_shutter_r
T = 1.0/Fs                      # sampling interval
t = np.arange(0,L)*T            # time vector

plt.subplot(2,1,1)
plt.plot(t,y,'k-')
plt.xlabel('time')
plt.ylabel('amplitude')

plt.subplot(2,1,2)
n = len(y)                       # length of the signal
k = np.arange(n)
T = n/Fs
frq = k/T # two sides frequency range
freq = frq[range(n/2)]           # one side frequency range

Y = np.fft.fft(y)/n              # fft computing and normalization
Y = Y[range(n/2)]

plt.plot(freq, abs(Y), 'r-')  
plt.xlim([0,1000])
```

```python
#fft method from luxapose
camera_rolling_shutter_r = 1/36000.0

y = withoutDC

Fs = 1/camera_rolling_shutter_r
T = 1/Fs
NFFT =  2**10  #should be the next power of 2 compared to the size of signal ex. len(signal)= 1944 so 2**11=2048
gain = 5

L = len(y)
t = numpy.arange(0,L)*T   #Return evenly spaced values within a given interval.

Y = numpy.fft.fft(y* gain, NFFT)/float(L)
f = Fs/2 * numpy.linspace(0,1,NFFT/2.0+1)
Y_plot = 2*abs(Y[0:NFFT/2.0+1])

peaks = scipy.signal.argrelmax(Y_plot)[0]
idx = numpy.argmax(Y_plot[peaks])
peak_freq = f[peaks[idx]]

print peak_freq
```
