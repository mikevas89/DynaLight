```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline


with open("/home/mike/AmbientLightValues.txt") as f:
    data = f.read()

data = data.split('\n')
del data[-1]
#print data
x = [row.split(' ')[0] for row in data]
y = [row.split(' ')[1] for row in data]

fig = plt.figure()

ax1 = fig.add_subplot(111)

ax1.set_title("AmbientLight over Time")    
ax1.set_xlabel('Time')
ax1.set_ylabel('Illuminance lx')

ax1.plot(x,y, c='r', label='AmbientLight')

leg = ax1.legend()

plt.show()

```
